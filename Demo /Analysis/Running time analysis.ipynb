```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_ip_30 = pd.read_csv('time-dat/IP_d30.csv', header=None)
df_heu_30 = pd.read_csv('time-dat/Heuristics_d30.csv', header=None)
df_csp_30 = pd.read_csv('time-dat/CSP_d30.csv', header=None)
```

```python
df_ip_30.columns = ['N', 'MIP']
df_ip_30.set_index('N', inplace=True)
df_ip_30['MIP'] = df_ip_30['MIP'].apply(np.log)
```

```python
df_heu_30.columns = ['N', 'Heuristics']
df_heu_30.set_index('N', inplace=True)
df_heu_30['Heuristics'] = df_heu_30['Heuristics'].apply(np.log)
```

```python
df_csp_30.columns = ['N', 'CSP']
df_csp_30.set_index('N', inplace=True)
df_csp_30['CSP'] = df_csp_30['CSP'].apply(np.log)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(df_ip_30.index, df_ip_30['MIP'], label='MIP (log)')
plt.plot(df_heu_30.index, df_heu_30['Heuristics'], label='Heuristics (log)')
plt.plot(df_csp_30.index, df_csp_30['CSP'], label='CSP (log)')
plt.legend(loc='upper left')
plt.suptitle('Running time comparison within a month (D = 30)')
plt.xlabel('Number of employees')
plt.ylabel('Running time (s)')
plt.savefig('d30.jpeg')
```
